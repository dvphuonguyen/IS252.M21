<h3>Loại bỏ các tag html trong file văn bản

```python
path = "E:\\IS252\\Lab1\\VNExpressSampleData"

os.chdir(path)

def cleanhtml(raw_html):
  cleanr = re.compile('<.*?>')
  cleantext = re.sub(cleanr, '', raw_html).strip()
  return cleantext
  
def read_text_file(file_path):
    with open(file_path, 'r', encoding = 'utf8') as f:
        file_data = f.read()
        file_data = cleanhtml(file_data)

        with open('E:\\IS252\\Lab1\\VNExpressSampleData\\result\\result.txt','a', encoding = 'utf8') as file:
            file.write(file_data)

for file in os.listdir():
    if file.endswith(".txt"):
        file_path = f"{path}/{file}"
  
        read_text_file(file_path)


with open('E:\\IS252\\Lab1\\VNExpressSampleData\\result\\result.txt', 'r', encoding = 'utf8') as f:
        print(f.read())
```
