<h3>Do điều chỉnh cảm biến nên dữ liệu đo được bắt đầu từ ngày 03/01/2018 mới chính xác. Những dữ liệu còn lại không có giá trị, êu cầu sinh viên loại bỏ

```python
df.head(12)
```

```python
df["time"] = pd.to_datetime(df["time"])
df = df[(df["time"] >= "1/3/2018")]
```

```python
df.head(12)
```

<h3>Từ cột time sinh viên bóc tách dữ liệu thành thời điểm đo đạc trong ngày tính theo phút, đặt tên là minutes

```python
df["minutes"] = df["time"].dt.hour*60 + df["time"].dt.minute + df["time"].dt.second/60
```

```python
df.head(12)
```

<h3>Vẽ đồ thị phân tán lần lượt biểu diễn dữ liệu nồng độ khí CO, độ ẩm và nhiệt độ theo cột minutes vừa tạo ở yêu cầu trên

<p>CO và minutes

```python
#importing library
import matplotlib.pyplot as plt
```

```python
df.info()
```

```python
plt.scatter(df["minutes"], df["co_level"], linewidths = 1, s = 100)
plt.title('Scatter 1')
plt.xlabel('minutes')
plt.ylabel('co_level')
plt.show()
```

<p>Biểu đồ trên cho thấy, nồng độ CO hầu như bằng 0 tại mọi thời điểm. Nên thuộc tính nồng độ CO không ảnh hưởng tới thuộc tính minutes

Độ ẩm và minutes

```python
plt.scatter(df["minutes"], df["humidity"], c ="pink",  linewidths = 1, s = 100)
plt.title('Scatter 2')
plt.xlabel('minutes')
plt.ylabel('humidity')
plt.show()
```

<p>Biểu đồ trên cho thấy, độ ẩm hầu như bằng 0 tại mọi thời điểm. Nên thuộc tính độ ẩm không ảnh hưởng tới thuộc tính minutes

Nhiệt độ và minutes

```python
plt.scatter(df["minutes"], df["temperature"], c ="green",  linewidths = 1, s = 100)
plt.title('Scatter 3')
plt.xlabel('minutes')
plt.ylabel('temperature')
plt.show()
```

<p>Biểu đồ trên cho thấy, nhiệt độ hầu như bằng 2 - 3 (1e6) tại mọi thời điểm. Nên thuộc tính nhiệt độ không ảnh hưởng tới thuộc tính minutes
